```python
import pandas as pd

df = pd.read_csv('ogloszenia_trojmiasto.csv')
df.head()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
df = df.dropna(subset=['price_total_zl', 'area_m2'])
df = df.drop(columns=['id'])
df.head()
```

```python
print(df.shape)
```

```python
df = df[df['price_total_zl']!='zapytaj o cenę']
df['price_total_zl_cleaned'] = df['price_total_zl'].astype(float)
```

```python
print(df.shape)
```

```python
df.info()
```

```python
# Tworzę zmienną z 'cleaned' w nazwie, bo potem mi się przyda zmienna z tą nazwą
df['price_per_m2_zl_cleaned'] = df['price_per_m2_zl'].astype(float)
```

```python
# Dodaję nową zmienną: area_per_room
df['area_per_room'] = df['area_m2'] / df['rooms'].astype(float).round(1)
df.head()
```

```python
# Sprawdzam, czy zostały jakieś NaNy
df.isna().sum()
```

```python
# Zapisuję wyczyszczony plik
df.to_csv('cleaned_trojmiasto.csv', index=False)
```

```python
# Polecenie: Przefiltruj dane dla Gdańska, Gdyni i Sopotu.
df_gdynia = df[df['city'] == 'Gdynia']
df_gdansk = df[df['city'] == 'Gdańsk']
df_sopot = df[df['city'] == 'Sopot']
print('Gdynia:', df_gdynia.shape, 'Gdańsk:', df_gdansk.shape, 'Sopot:', df_sopot.shape, 'Total:', df.shape)
```
